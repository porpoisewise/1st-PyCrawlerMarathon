```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

'''
Your Code
'''
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

#response


data = json.loads(response)

data_count = 0

print('å›å‚³ç­†æ•¸' + str(len(data)))

dict_data = data[0]
print('åŒ…å«æ¬„ä½')
print(dict_data.keys())
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

'''
Your Code
'''

for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

'''
Your Code
'''
import numpy as np

hot_info_commentCount = []
hot_info_likeCount = []
info_commentCount = []
info_likeCount = []

for d in data:
    #print(d['commentCount'],d['likeCount'])
    if d['commentCount'] > 10 : # ç†±é–€
        hot_info_commentCount.append(d['commentCount'])
        hot_info_likeCount.append(d['likeCount'])
    else:
        info_commentCount.append(d['commentCount'])
        info_likeCount.append(d['likeCount'])

print('--- ç†±é–€ ---')
print('å¹³å‡ç•™è¨€äººæ•¸' + str(np.mean(hot_info_commentCount)))
print('å¹³å‡æŒ‰è®šäººæ•¸' + str(np.mean(hot_info_likeCount)))

print('--- éç†±é–€ ---')
print('å¹³å‡ç•™è¨€äººæ•¸' + str(np.mean(info_commentCount)))
print('å¹³å‡æŒ‰è®šäººæ•¸' + str(np.mean(info_likeCount)))
```
