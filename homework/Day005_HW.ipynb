1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

'''
Your Code
'''
import requests
import json

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text

#response


data = json.loads(response)

data_count = 0

print('回傳筆數' + str(len(data)))

dict_data = data[0]
print('包含欄位')
print(dict_data.keys())
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

'''
Your Code
'''

for d in data:
    print(d['title'])
    print(d['createdAt'])
    print(d['commentCount'])
    print(d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

'''
Your Code
'''
import numpy as np

hot_info_commentCount = []
hot_info_likeCount = []
info_commentCount = []
info_likeCount = []

for d in data:
    #print(d['commentCount'],d['likeCount'])
    if d['commentCount'] > 10 : # 熱門
        hot_info_commentCount.append(d['commentCount'])
        hot_info_likeCount.append(d['likeCount'])
    else:
        info_commentCount.append(d['commentCount'])
        info_likeCount.append(d['likeCount'])

print('--- 熱門 ---')
print('平均留言人數' + str(np.mean(hot_info_commentCount)))
print('平均按讚人數' + str(np.mean(hot_info_likeCount)))

print('--- 非熱門 ---')
print('平均留言人數' + str(np.mean(info_commentCount)))
print('平均按讚人數' + str(np.mean(info_likeCount)))
```
