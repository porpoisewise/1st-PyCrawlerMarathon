```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
'''
Your Code
'''
list_all = table.find_all('td')
d = {}
key = ''
date = ''
for dtemp in list_all:    
    if 'style' in dtemp.attrs:        
        #print(dtemp.text)
        key = dtemp.text
        d[key] = {}
    else:
        if 'class' not in dtemp.attrs:
            #print(dtemp.text)
            if '/' in dtemp.text:
                date = dtemp.text
            elif '\xa0' in dtemp.text:
                continue
            else:
                d[key][date] = dtemp.text
                
d
```
